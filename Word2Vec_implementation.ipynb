```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
nltk.download('stopwords')
nltk.download('punkt_tab')
nltk.download('wordnet')
```

```python
messages = pd.read_csv('./ham_spam_ds.csv')
print(messages.shape), messages.head()
```

```python
english_stopwords = stopwords.words('english')
english_stopwords.remove('not')
lemmatizer = WordNetLemmatizer()

def fetch_words(message):
    cleaned_message = re.sub(r'[^a-zA-Z0-9\s]', ' ', message)
    words = word_tokenize(cleaned_message.lower())
    lemmatized_tokens = [lemmatizer.lemmatize(token, pos='v') for token in words]
    return lemmatized_tokens

processed_words = messages['message'].apply(fetch_words).tolist()
processed_words
```

```python
model = Word2Vec(processed_words, epochs=100, vector_size=300)
model
```

```python
model.wv.index_to_key
```

```python
model.corpus_count
```

```python
model.wv.similar_by_word('we')
```

```python
def avgWord2Vec(doc):

  vectors = []
  for word in doc:
    if word in model.wv.index_to_key:
      vectors.append(model.wv[word])
    else:
      vectors.append(np.zeros(300,))

  return np.mean(vectors, axis=0)
```

```python
avgWord2Vec(processed_words[0])
```

```python
!pip install tqdm
from tqdm.auto import tqdm
tqdm.pandas(desc="Processing...")
```

```python
processed_messages = []
for index in tqdm(range(len(processed_words))):
  processed_messages.append(avgWord2Vec(processed_words[index]))
processed_messages = np.array(processed_messages)
```

```python
X = pd.DataFrame(processed_messages)
X.head()
```

```python
# processed_messages -> X
y = pd.get_dummies(messages['label'], dtype=int)[['spam']]
y[:10]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.20)
X_train.shape, X_test.shape, messages.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

spam_detect_model = RandomForestClassifier().fit(X_train, y_train)
```

```python
y_pred = spam_detect_model.predict(X_test)
y_pred
```

```python
print(accuracy_score(y_test, y_pred), '\n', classification_report(y_test, y_pred))
```

```python
def is_text_spam(text):

  words = fetch_words(text)
  vector = avgWord2Vec(words)
  is_spam = spam_detect_model.predict([vector])
  if (is_spam):
    return 'spam'
  return 'ham'

is_text_spam("how is it eveything?")
```
